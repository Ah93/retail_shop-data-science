# Feature Engineering

```python
data["profit"] = data["quantity"] * data["price"] #create profit feature
```

```python
# Ensure order_date is datetime
data['order_date'] = pd.to_datetime(data['order_date'])

# Sort by date to ensure correct lag calculation
data = data.sort_values('order_date')

# Create lag features
data['last_month_profit'] = data['profit'].shift(1)  # previous row's profit
data['avg_last_3_months_profit'] = data['profit'].rolling(window=3).mean().shift(1)

# Month-over-month change
data['month_over_month_change'] = data['profit'].pct_change()

# Cumulative profit till that date
data['cumulative_sales_to_date'] = data['profit'].cumsum()

# Add season column (Winter = 1, Spring = 2, Summer = 3, Fall = 4)
def assign_season(month):
    if month in [12, 1, 2]:
        return 1  # Winter
    elif month in [3, 4, 5]:
        return 2  # Spring
    elif month in [6, 7, 8]:
        return 3  # Summer
    else:
        return 4  # Fall

data['season'] = data['order_date'].dt.month.apply(assign_season)
```

```python
# data['order_date'] = pd.to_datetime(data['order_date'])
data['order_month'] = data['order_date'].dt.month
data['order_day'] = data['order_date'].dt.day
data['order_weekday'] = data['order_date'].dt.weekday
data['order_year'] = data['order_date'].dt.year
```

```python
import pandas as pd
import joblib
from calendar import month_name

# === Assume you already have a DataFrame `data` loaded ===
# If not, load it like:
# data = pd.read_csv("your_data.csv")  # Adjust as needed

# Extract lookup tables
product_encoded_table = data[['product_name', 'product_encoded']].drop_duplicates()
city_encoded_table = data[['city', 'city_encoded']].drop_duplicates()

# Helper input functions
def get_float(prompt):
    return float(input(prompt))

def get_int(prompt):
    return int(input(prompt))

def get_encoded_value_from_table(prompt, column_name, table):
    while True:
        user_input = input(prompt).strip().lower()
        match = table[table[column_name].str.lower() == user_input]
        if not match.empty:
            return match.iloc[0, 1]
        else:
            print(f"'{user_input}' not found in {column_name}. Try again. Examples: {', '.join(table[column_name].unique()[:5])}")

# === User Input for Latest Row ===
print("Enter details for the latest known row:")
quantity = get_float("Quantity: ")
price = get_float("Price: ")
discount = get_float("Discount (%): ")  # New discount field
discounted_price = price * (1 - discount / 100)

product_encoded = get_encoded_value_from_table("Enter product name: ", "product_name", product_encoded_table)
city_encoded = get_encoded_value_from_table("Enter city name: ", "city", city_encoded_table)

latest_row = {
    'quantity': quantity,
    'price': discounted_price,
    'product_encoded': product_encoded,
    'city_encoded': city_encoded,
    'last_month_profit': get_float("Last Month Profit: "),
    'avg_last_3_months_profit': get_float("Average Last 3 Months Profit: "),
    'month_over_month_change': get_float("Month over Month Change: "),
    'cumulative_sales_to_date': get_float("Cumulative Sales to Date: "),
    'season': get_int("Season (e.g. 1=Winter, 2=Spring...): "),
    'order_month': get_int("Order Month (1-12): "),
    'order_day': get_int("Order Day: "),
    'order_weekday': get_int("Order Weekday (0=Mon): "),
    'order_year': get_int("Order Year: "),
}

# Load model and scaler
model = joblib.load("sales_forecast_model.pkl")
scaler = joblib.load("sales_forecast_scaler.pkl")

# === First Forecast ===
input_df = pd.DataFrame([latest_row])
scaled_input = scaler.transform(input_df)
first_pred = model.predict(scaled_input)[0]
forecast_month = latest_row['order_month']
print(f"\nProfit forecast for {month_name[forecast_month]} ({forecast_month}): {first_pred:.2f}")

# === Second Month Forecast (1-month ahead) ===
next_row = latest_row.copy()
next_row['order_month'] = (forecast_month % 12) + 1
next_row['order_year'] += 1 if next_row['order_month'] == 1 else 0
next_row['last_month_profit'] = first_pred
next_row['avg_last_3_months_profit'] = (latest_row['avg_last_3_months_profit'] * 2 + first_pred) / 3
next_row['month_over_month_change'] = (first_pred - latest_row['last_month_profit']) / latest_row['last_month_profit']
next_row['cumulative_sales_to_date'] = latest_row['cumulative_sales_to_date'] + first_pred
# Adjust season based on new month
next_row['season'] = ((next_row['order_month'] % 12 + 3) // 3)

next_df = pd.DataFrame([next_row])
next_scaled = scaler.transform(next_df)
second_pred = model.predict(next_scaled)[0]
print(f"Profit forecast for {month_name[next_row['order_month']]} ({next_row['order_month']}): {second_pred:.2f}")
```
